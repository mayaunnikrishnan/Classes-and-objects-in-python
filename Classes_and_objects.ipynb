## **Inheritance in Python**
capability of one class to derive or inherit the properties from another class.

Benefits of inheritance are:

@ real-world relationships

@ reusability of a code

@ add more features to a class without modifying it

@ transitive in nature, which means that if class B inherits from another class A, then all the subclasses of
B would automatically inherit from class A.

**Creating a Parent Class/Super class/Base class**

Parent class is the class being inherited from,also called base class.
Any class can be a parent class, so the syntax is same as that of creating a class.


**Creating a Child Class/sub class**

Python Inheritance Syntax



```
Class BaseClass:
     {Body}
Class DerivedClass(BaseClass):
      {Body}
```



### **Different types of Inheritance:**
@ **Single inheritance**: When a child class inherits from only one parent class, it is called single inheritance.

@ **Multiple inheritances**: When a child class inherits from multiple parent classes, it is called multiple
inheritances.

@ **Multilevel Inheritance** :features of the base class and the derived class are further inherited into the new
derived class.

@ **Hierarchical Inheritance**: When more than one derived class are created from a single base class

@ **Hybrid Inheritance**: Inheritance consisting of multiple types of inheritance is called hybrid inheritance.

````python
# single inheritance
class parent:
   def display(self):
      print("im a parent class")
class child(parent):
   def show(self):
      print("im a child class")
c=child()
c.display()
c.show()
````

````python
# multiple inheritance
class Father:
    def display(self):
        print("I'm father class")

class Mother:
    def display2(self):
        print("I'm a mother class")

class Child(Father, Mother):
    def show(self):
        print("I'm a child class")

# Creating an instance of the Child class
c = Child()

# Calling methods
c.display()   # Output: I'm father class
c.display2()  # Output: I'm a mother class
c.show()      # Output: I'm a child class
````

````python
#multilevel inheritance
class Grandpa:
    def greatdisplay(self):
        print("I'm grandpa class")

class Father(Grandpa):
    def display(self):
        print("I'm father class")

class Child(Father):
    def show(self):
        print("I'm child class")

c = Child()
c.greatdisplay()  # Output: I'm grandpa class
````

````python
# heirarchial inheritance
class Father:
    def display(self):
        print("I'm father class")

class Boy(Father):
    def show1(self):
        print("I'm boy child class")

class Girl(Father):
    def show2(self):
        print("I'm girl child class")

b = Boy()
g = Girl()
b.display()  # Output: I'm father class
````

````python
# hybrid inheritance
class Grandpa:
    def greatdisplay(self):
        print("I'm great grandfather class")

class Father(Grandpa):
    def display1(self):
        print("I'm father class")

class Mother(Grandpa):
    def display2(self):
        print("I'm mother class")

class Child(Father, Mother):
    def show(self):
        print("I'm a child class")

c = Child()
c.greatdisplay()  # Output: I'm great grandfather class
c.display1()      # Output: I'm father class
c.display2()      # Output: I'm mother class
c.show()          # Output: I'm a child class
````

## **Encapsulation in Python**
Encapsulation is one of the fundamental concepts in object-oriented programming (OOP).
It describes the
idea of wrapping data and the methods that work on data within one unit. This puts restrictions on
accessing variables and methods directly and can prevent the accidental modification of data.

**Protected members**

Protected members are those members of the class that cannot be accessed outside the class but can be
accessed from within the class and its subclasses

**Private members**

Private members are similar to protected members, the difference is that the class members declared
private should neither be accessed outside the class nor by any base class

**Protected members of Parent class**

Although the protected variable can be accessed out of the class as well as in the derived class(modified
too in derived class), it is customary(convention not a rule) to not access the protected out the class body

````python
class Parent:
    def display(self):
        self.__a = 2
        print("I'm a parent class")
        print("Parent protected 'a' is", self.__a)

class Child(Parent):
    def show(self):
        print("I'm a child class")
        self.__a = 3  # Modifying the protected member of the parent class
        print("Child protected 'a' is", self.__a)
````

## **Polymorphism in Python**

Polymorphism is a fundamental concept in object-oriented programming that allows objects of different classes to be treated as objects of a common superclass. It enables you to write code that can work with objects of different types in a unified way.

**There are two main types of polymorphism**:

compile-time (**method overloading**)

runtime (**method overriding**).

### Compile-Time Polymorphism (Method Overloading):

In some programming languages (but not in Python), you can define multiple methods with the same name but different parameter lists. This is known as method overloading. Python doesn't support method overloading in the traditional sense.

### Runtime Polymorphism (Method Overriding):
This is the more common form of polymorphism in Python. It involves overriding a method in a derived class to provide a specific implementation while maintaining the method signature of the base class.

````python
#example demonstrating a form of compile-time polymorphism in Python:
class Calculator:
    def add(self, a, b=None):
        if b is None:
            return a
        else:
            return a + b

# Creating an instance of Calculator
calc = Calculator()

# Using the add method with different arguments
result1 = calc.add(5)
result2 = calc.add(5, 10)

print(result1)  # Output: 5
print(result2)  # Output: 15
````

In this example, the add method in the Calculator class can accept one or two arguments. If only one argument is provided, it returns that value. If two arguments are provided, it returns their sum. This demonstrates a form of compile-time polymorphism, where the behavior of the method varies based on the number of arguments passed to it.

````python
#example of method overriding (runtime polymorphism) in Python:
class Animal:
    def make_sound(self):
        pass  # Abstract method, to be overridden in subclasses

class Dog(Animal):
    def make_sound(self):
        return "Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

# Polymorphism in action
def animal_sound(animal):
    return animal.make_sound()

# Creating instances
dog = Dog()
cat = Cat()

# Using the same function for different objects
print(animal_sound(dog))  # Output: Woof!
print(animal_sound(cat))  # Output: Meow!
````

In this example:

Animal is the base class with an abstract method make_sound, which
should be overridden in subclasses.

Dog and Cat are derived classes that override the make_sound method to provide their own implementations.

The function animal_sound takes an instance of Animal (or any subclass) and calls its make_sound method.

This demonstrates polymorphism, where the same function is used with different types of objects, resulting in different behavior based on their individual implementations.

In summary, polymorphism allows you to write more flexible and reusable code by treating different objects in a unified manner, even if they belong to different classes.


## **Abstarctmethod**

@ **Abstact class**: which have abstact methods

@ **Abstract method**: method with declaration not the definition
object cannot be instantiated for abstract class

@**Concrete class**: class without abstact method
object can only be instantiated for concete class



````python
from abc import ABC, abstractmethod

class MyAbstractClass(ABC):
    @abstractmethod
    def display(self):
        pass

    @abstractmethod
    def show(self):
        pass

class Demo1(MyAbstractClass):
    def display(self):
        print("Display method in Demo1")

class Demo2(MyAbstractClass):
    def display(self):
        print("Display method in Demo2")

    def show(self):
        print("Show method in Demo2")

d2 = Demo2()
d2.display()
d2.show()
````

In this corrected code:

ABC and abstractmethod are imported from the abc module. The ABC class is used as a base class for creating abstract base classes, and the abstractmethod decorator is used to define abstract methods.

MyAbstractClass is defined as an abstract base class with abstract methods display and show.

Demo1 and Demo2 are concrete classes that inherit from MyAbstractClass.

Demo1 provides an implementation for the display method, while Demo2 provides implementations for both the display and show methods.

#Magical Methods in Python
**Overview**

Magical methods, also known as "dunder" (double underscore) methods, are special methods in Python that have double underscores at the beginning and end of their names (e.g., __init__, __str__). These methods allow you to define how objects of your class behave with built-in functions and operators. They are a key part of making your custom classes integrate smoothly with Pythonâ€™s syntax and operations.

**Common Magical Methods**

Here are some of the most commonly used magical methods:

**Object Initialization and Representation**

__init__(self, ...): This method is called when an instance of the class is created. It initializes the object's attributes.

````python
class MyClass:
    def __init__(self, value):
        self.value = value

obj = MyClass(42)
print(obj.value)  # Output: 42
````

`__str__(self)`:

This method is called by the str() function and the print statement to get a string representation of the object.

````python
class MyClass:
    def __str__(self):
        return f'MyClass with value: {self.value}'
````

`__repr__(self)`:

This method is called by the repr() function and is used to get a detailed string representation of the object, often one that could be used to recreate the object.

````python
class MyClass:
    def __repr__(self):
        return f'MyClass({self.value})'
````

````python
class MyClass:
    def __mul__(self, other):
        return MyClass(self.value * other.value)
````
